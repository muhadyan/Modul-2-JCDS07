# Latihan 1: Grafik Persamaan Linear

3x + y = 15 <br>
x + y = 7 <br>
Berapa nilai x dan y?

```python
x = np.array([[3, 1],[1, 1]])
y = np.array([15, 7])
hasil = np.linalg.solve(x,y)
hasil
```

```python
hasilX = round(hasil[0])
hasilY = round(hasil[1])
print(hasilX)
print(hasilY)
```

1. 3x + y = 15 <br>
    -titik potong sb x => y = 0 => 3x = 15

```python
y1 = 0
x1 = np.linalg.solve([[3]], [15])[0]
x1
```

    -titik potong sb y = 0 => y = 15

```python
x2 = 0
y2 = np.linalg.solve([[1]], [15])[0]
y2
```

2. x + y = 7 <br>
    -titik potong sb x => y = 0 => x = 7

```python
y3 = 0
x3 = np.linalg.solve([[1]], [7])[0]
x3
```

    -titik potong sb y => x = 0 => y = 7

```python
x4 = 0
y4 = np.linalg.solve([[1]], [7])[0]
y4
```

```python
plt.plot([x1, y1], [x2, y2])
plt.plot([x3, y3], [x4, y4])
plt.plot(hasilX, hasilY, 'ro')
plt.annotate(
    'Titik\npotong\n(4,3)', xy=(hasilX, hasilY),
    xytext=(5,4), arrowprops = {'facecolor':'r', 'shrink':.1}
)
plt.grid(True)
plt.show()
```

## Soal 1

```python
def saham(x):
    df = pd.read_csv(x, parse_dates= ['Date'], index_col= 'Date')
    df = df.resample('D').sum()
    df.iloc[:, 0:5] = df.iloc[:, 0:5].replace(0, np.NaN)
    df = df.fillna(method='ffill')
    return df

EXCL = saham('EXCL.JK.csv')
FREN = saham('FREN.JK.csv')
ISAT = saham('ISAT.JK.csv')
TLKM = saham('TLKM.JK.csv')
```

```python
waktu = EXCL.iloc[:].index.tolist()

plt.plot(waktu, EXCL['Close'], color='g' ,label='XL')
plt.plot(waktu, FREN['Close'], color='lightblue', label='Smartfren')
plt.plot(waktu, ISAT['Close'], color='b', label='Indosat')
plt.plot(waktu, TLKM['Close'], color='r', label='Telkomsel')

plt.title('Harga Historis Saham Provider Telco Indonesia')
plt.xlabel('Tanggal')
plt.ylabel('Rupiah (IDR)')
plt.grid(True)
plt.legend()

plt.show()
```
