```python
import csv
import json
import xlrd
import mysql.connector
import pymongo
import pandas as pd
```

## 1. Convert CSV into Data Frame

```python
with open('data.csv', 'r') as x:
    baca = csv.reader(x)
    z = list(baca)
col = list(z[0])
data = list(z[1:])
df = pd.DataFrame(data, columns=col)
df
```

## 2. Convert JSON into Data Frame

```python
with open('data.json', 'r') as x:
    y = json.load(x)
col = list(y[0].keys())
data = []
for i in y:
    data.append(list(i.values()))
df = pd.DataFrame(data, columns=col)
df
```

## 3. Convert XLSX into Data Frame

```python
buka = xlrd.open_workbook('data.xlsx')
sheet = buka.sheet_by_index(0)

data = []
for i in range(sheet.nrows):
    data.append(sheet.row_values(i))

df = pd.DataFrame(data[1:], columns=data[0])
df
```
