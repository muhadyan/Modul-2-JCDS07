# II. Bar Chart

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})
```

## A. Making the bar chart

```python
plt.bar(df['x'], df['y'])
plt.show()
```

## B. Resize label's font

```python
plt.bar(df['x'], df['y'])
plt.xlabel('Ini X', fontsize=30)
plt.ylabel('Ini Y')
plt.show()
```

## C. Modif Title

```python
plt.bar(df['x'], df['y'])
plt.title('Ini bar chart', fontdict={
    'family':'Impact', 'size': 30, 'color':'lightgreen'
})
plt.show()
```

## D. Edit bar

### 1. Satu warna

```python
plt.bar(df['x'], df['y'],
       color='pink',
        edgecolor='black', linewidth=10)
plt.show()
```

### 2. Banyak warna

```python
plt.bar(df['x'], df['y'],
       color=['r','g','b','y','k'],
        edgecolor='black', linewidth=10)
plt.show()
```

### 3. Horizontal Bar

```python
plt.barh(df['x'], df['y'],
       color=['r','g','b','y','k'],
        edgecolor='black', linewidth=10)
plt.show()
```

## E. Tampilin Value

```python
plt.bar(df['x'], df['y'],
       color=['r','g','b','y','k'],
        edgecolor='black', linewidth=5)
plt.yticks(np.arange(0,12,1))

for i in df['y']:
    plt.text(df[df['y'] == i]['x']-.075, i+.2, i,
            fontdict={'size':20})

plt.show()
```

## F. Multiple Bar

### 1. Yg pendek di bawah

```python
plt.bar(df['x'], df['y'],
       color='k',
        edgecolor='black', linewidth=5)
plt.bar(df['x'], df['z'],
       color='pink',
        edgecolor='black', linewidth=5)

plt.yticks(np.arange(0,12,1))

for i in df['y']:
    plt.text(df[df['y'] == i]['x']-.075, i+.2, i,
            fontdict={'size':20})

plt.show()
```

### 2. Yg tinggi di bawah (ditumpuk pake bottom)

```python
plt.bar(df['x'], df['z'],
       color='pink',
        edgecolor='black', linewidth=5, bottom=df['y'])
plt.bar(df['x'], df['y'],
       color='k',
        edgecolor='black', linewidth=5)

plt.yticks(np.arange(0,12,1))

for i in df['y']:
    plt.text(df[df['y'] == i]['x']-.075, i+.2, i,
            fontdict={'size':20})

plt.show()
```

## G. Menampilkan Nilai Error

```python
plt.bar(df['x'], df['y'],
       color='pink',
       edgecolor='black', linewidth=10,
       yerr=0.5)

plt.yticks(np.arange(0,12,1))
plt.show()
```

# IV. Combine Plot, Scatter, & Bar

'zorder' untuk penempatan masing-masing cartnya

```python
plt.scatter(
    [1,2,3], [3,6,2],
    marker='*', s=300, color='lightblue', edgecolor='g',
    zorder=3
)
plt.plot([1,2,3], [3,6,2], zorder=2, color='r')
plt.bar([1,2,3], [3,6,2], zorder=1)
plt.show()
```

# V. Pie Chart

## A. Create

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
plt.pie(
    persentase, labels= keterangan
)
plt.show()
```

## B. Edit

### 1. Warna Pie

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
plt.pie(
    persentase, labels= keterangan,
    colors= ['pink', 'black']
)
plt.show()
```

### 2. Start Angle

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
plt.pie(
    persentase, labels= keterangan,
    colors= ['pink', 'black'],
    startangle = 45
)
plt.show()
```

### 3. Arahnya jadi Jarum jam

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
plt.pie(
    persentase, labels= keterangan,
    colors= ['pink', 'black'],
    startangle = 45,
    counterclock=False
)
plt.show()
```

### 4. Property Text

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
plt.pie(
    persentase, labels= keterangan,
    colors= ['pink', 'black'],
    startangle = 45,
    counterclock=False,
    textprops={'size':18, 'color':'g'}
)
plt.show()
```

### 5. Tampilin Value

#### a. Shows

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
plt.pie(
    persentase, labels= keterangan,
    colors= ['pink', 'black'],
    startangle = 45,
    counterclock=False,
    textprops={'size':18, 'color':'g'},
    autopct = '%1.2f%%'  # untuk tampilin float dengan 2 angka di belakang koma. Int juga bisa.
)
plt.show()
```

#### b. Ngeubah warnanya

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
x, y, z = plt.pie(
    persentase, labels= keterangan,
    colors= ['pink', 'black'],
    startangle = 45,
    counterclock=False,
    textprops={'size':18, 'color':'g'},
    autopct= '%1.2f%%'
)
for i in z:
    i.set_color('w')
plt.show()
```

### 6. Atur besarnya pie

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
plt.pie(
    persentase, labels= keterangan,
    colors= ['pink', 'black'],
    startangle = 45,
    counterclock=False,
    textprops={'size':18, 'color':'g'},
    autopct= '%1.2f%%',
    radius= 3
)
plt.show()
```

### 7. Ngeluarin sebagian data

```python
persentase = [25, 55, 20]
keterangan = ['Pak A', 'Pak B', 'Pak C']
x, y, z = plt.pie(
    persentase, labels= keterangan,
    colors= ['g', 'r', 'b'],
    startangle = 45,
    counterclock=False,
    textprops={'size':18, 'color':'g'},
    autopct= '%1.2f%%',
    explode=(0,0,.1)
)

for i in z:
    i.set_color('w')
plt.show()
```

### 8. Tampilin Shadow

```python
persentase = [25, 75]
keterangan = ['Pak A', 'Pak B']
plt.pie(
    persentase, labels= keterangan,
    colors= ['pink', 'black'],
    startangle = 45,
    counterclock=False,
    shadow=True
)
plt.show()
```

# VI. Histogram

## A. Create

```python
usia = [22,21,31,42,32,45,65,55,43,22,21,34,43,44]
par = [15,20,25,30,35,40,45,50,55,60,65]
plt.hist(usia, par)
plt.xticks(np.arange(15, 66, 5))
plt.show()
```

## B. Ngeubah tipe histogramnya

### 1 Variabel

```python
usia = [22,21,31,42,32,45,65,55,43,22,21,34,43,44]
par = [15,20,25,30,35,40,45,50,55,60,65]
plt.hist(usia, par, histtype='stepfilled')
plt.xticks(np.arange(15, 66, 5))
plt.show()
```

### 2 Variabel

```python
usia = [22,21,31,42,32,45,65,55,43,22,21,34,43,44]
usia2 = [33,31,41,21,22,25,25,55,53,52,41,44,23,24]
par = [15,20,25,30,35,40,45,50,55,60,65]
plt.hist([usia, usia2], par, histtype='barstacked')
plt.xticks(np.arange(15, 66, 5))
plt.show()
```
