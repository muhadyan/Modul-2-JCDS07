```python
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
```

```python
dbku = mysql.connector.connect(
    host = "localhost",
    user = "Adyan",
    passwd = "C3ndolbig",
    database = "world"
)
```

```python
query = "select * from Country where Region = 'Southeast Asia' order by Name"
df = pd.read_sql(query, dbku)
df
```

```python
plt.style.use('seaborn')

plt.figure('ASEAN3', figsize=(15,8))
plt.bar(df.Name, df.Population, color=['g', 'orange', 'g', 'r', 'purple', 'brown', 'pink', 'grey', 'y', 'lightblue', 'b'])
plt.title('Populasi Negara ASEAN', fontsize=20)
plt.xlabel('Negara')
plt.ylabel('Populasi (x100jt jiwa)')
plt.xticks(rotation= 45)

for i in df['Population']:
    plt.text(df[df['Population'] == i]['Name'], i, i,
            fontdict={'size':13})

plt.show()
```

```python
plt.figure('ASEAN', figsize=(8,8))
x, y, z = plt.pie(df['Population'], labels= df['Name'], autopct = '%1.1f%%')
for i in z:
    i.set_color('w')
plt.title('Persentase Penduduk ASEAN', fontsize=20)

plt.show()
```

```python
plt.figure('ASEAN3', figsize=(15,8))
plt.bar(df.Name, df.GNP, color=['g', 'orange', 'g', 'r', 'purple', 'brown', 'pink', 'grey', 'y', 'lightblue', 'b'])
plt.title('Pendapatan Bruto Nasional ASEAN', fontsize=20)
plt.xlabel('Negara')
plt.ylabel('Gross National Product (US$)')
plt.xticks(rotation= 45)

for i in df['GNP']:
    plt.text(df[df['GNP'] == i]['Name'], i, i,
            fontdict={'size':13})

plt.show()
```

```python
plt.figure('ASEAN4', figsize=(8,8))
x, y, z = plt.pie(df['SurfaceArea'], labels= df['Name'], autopct = '%1.1f%%')
for i in z:
    i.set_color('w')
plt.title('Persentase Luas Daratan ASEAN', fontsize=20)

plt.show()
```
