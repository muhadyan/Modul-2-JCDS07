Cari data desa BPS: <br>
- Provinsi dengan desa terbanyak di th 03, 05, 08, 11, 14, 18 <br>
- Provinsi dengan desa tersedikit di th 03, 05, 08, 11, 14, 18

<b>How To:</b>
1. Buang header dan footer yg ga perlu
2. Jadikan provinsi sebagai primary key
3. Urus data yg kosong
4. Gabung desa tepi laut dan bukan tepi laut
5. Cari max di masing-masing tahun
6. Cari min di masing-masing tahun

```python
import pandas as pd
```

```python
df = pd.read_excel('data-bps.xlsx', header=3, skipfooter=1, na_values='-')
df = df.rename(columns = {'Unnamed: 0': 'Provinsi'})
df = df.set_index('Provinsi')
df
```

```python
for i in df:
    if i == '2003.1':
        break
    df[i] = df[i] + df[str(i)+'.1']
df
```

```python
for i in df:
    if i == '2003.1':
        break
    print(f'Desa terbanyak pada tahun {i} ada di Provinsi {(df.loc[df[i] == df[i].max()].index.tolist())[0]}')
```

```python
for i in df:
    if i == '2003.1':
        break
    print(f'Desa tersedikit pada tahun {i} ada di Provinsi {(df.loc[df[i] == df[i].min()].index.tolist())[0]}')
```
