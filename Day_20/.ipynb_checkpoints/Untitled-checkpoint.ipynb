```python
import folium
```

## A. Create

### 1. Tampilin Mapnya

```python
map = folium.Map(
    location = [-6.1754374,106.8272385])
map
```

### 2. Ngezoom

```python
map = folium.Map(
    location = [-6.1754374,106.8272385],
    zoom_start = 17
)
map
```

### 3. Model mapnya

```python
map = folium.Map(
    location = [-6.1754374,106.8272385],
    zoom_start = 17,
    tiles = 'Stamen Toner'
)
map
```

```python
map = folium.Map(
    location = [-6.1754374,106.8272385],
    zoom_start = 20,
    tiles = 'OpenStreetMap'
)

folium.Marker(
    location = [-6.1754374,106.8272385],
    tooltip = 'Halo',
    popup = '<b>MONAS</b><br><small>Jakarta</small>',
    icon = folium.Icon(
        color='orange',
        icon='cloud')
).add_to(map)

folium.Marker(
    location = [-6.1752374,106.8274385],
    tooltip = '<img width="100px" src="https://2.bp.blogspot.com/-0WirdbkDv4U/WxUkajG0pAI/AAAAAAAADNA/FysRjLMqCrw_XkcU0IQwuqgKwXaPpRLRgCLcBGAs/s1600/1528109954774.jpg">',
    popup = '<iframe width="300" height="200" src="https://www.youtube.com/embed/SEGbYdUe9jM" frameborder="0" allow="accelerometer; autoplay; encrypted-media; gyroscope; picture-in-picture" allowfullscreen></iframe>',
    icon = folium.Icon(
        color='purple',
        prefix='fa',
        icon='bullseye')
).add_to(map)

folium.Circle(
    location = [-6.1754374,106.8272385],
    radius = 100,
    color = 'red', fill = True, fill_color = 'yellow'
).add_to(map)

map.save('petaku.html')
map
```
