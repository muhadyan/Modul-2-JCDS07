## 1. Dengan menggunakan numpy, buatlah sebuah program python yang dapat menentukan nilai x, y & z dari 3 buah persamaan di atas! Hasil yang diharapkan saat program dieksekusi adalah:

```python
a = np.array([[1,-2,1], [3,1,-2], [7,-6,-1]])
b = np.array([6,4,10])
hasil = np.linalg.solve(a,b)
x = round(hasil[0])
y = round(hasil[1])
z = round(hasil[2])

print(f'Nilai x = {x}')
print(f'Nilai y = {y}')
print(f'Nilai z = {z}')
```

## 2. Dengan menggunakan matplotlib, buatlah sebuah program python yang dapat menggambarkan area 3D yang merepresentasikan masing-masing persamaan di atas. Hasil yang diharapkan saat program dieksekusi berupa 1 buah figure dengan 3 buah plot 3D sebagai berikut:

### x  - 2y +  z =  6

```python
z1 = 0
y1 = 0
x1 = np.linalg.solve([[1]], [6])[0]

z2 = 0
x2 = 0
y2 = np.linalg.solve([[-2]], [6])[0]

y3 = 0
x3 = 0
z3 = np.linalg.solve([[1]], [6])[0]

print(x1)
print(y2)
print(z3)
```

### 3x +  y - 2z =  4

```python
z4 = 0
y4 = 0
x4 = np.linalg.solve([[3]], [4])[0]

z5 = 0
x5 = 0
y5 = np.linalg.solve([[1]], [4])[0]

y6 = 0
x6 = 0
z6 = np.linalg.solve([[-2]], [4])[0]

print(x4)
print(y5)
print(z6)
```

### 7x - 6y -  z = 10

```python
z7 = 0
y7 = 0
x7 = np.linalg.solve([[7]], [10])[0]

z8 = 0
x8 = 0
y8 = np.linalg.solve([[-6]], [10])[0]

y9 = 0
x9 = 0
z9 = np.linalg.solve([[-1]], [10])[0]

print(x7)
print(y8)
print(z9)
```

```python
plt.figure('3D SPLTV', figsize=(20,20))

p = plt.subplot(331, projection='3d')
p.plot_wireframe([x1,y1,z1], [x2,y2,z2], np.array([x3,y3,z3]).reshape(1,-1))
p.set_xlabel('Nilai X')
p.set_ylabel('Nilai Y')
p.set_zlabel('Nilai Z')
plt.title('x  - 2y +  z =  6')

q = plt.subplot(332, projection='3d')
q.plot_wireframe([x4,y4,z4], [x5,y5,z5], np.array([x6,y6,z6]).reshape(1,-1))
q.set_xlabel('Nilai X')
q.set_ylabel('Nilai Y')
q.set_zlabel('Nilai Z')
plt.title('3x +  y - 2z =  4')

r = plt.subplot(333, projection='3d')
r.plot_wireframe([x7,y7,z7], [x8,y8,z8], np.array([x9,y9,z9]).reshape(1,-1))
r.set_xlabel('Nilai X')
r.set_ylabel('Nilai Y')
r.set_zlabel('Nilai Z')
plt.title('7x - 6y -  z = 10')

plt.show()
```
